```python
import os
import glob
from tqdm import tqdm_notebook as tqdm
import librosa
import IPython.display as ipd
import re

def play(aud, rate=16000):
    ipd.display(ipd.Audio(aud, rate=rate))
```

```python
TRANSCRIPT_PATH = 'Transcripts/'
AUDIO_PATH = 'Audios/'
OUTPUT_PATH = 'ChirppCollected/'

SAMPLERATE = 16000
```

```python
for i in tqdm(glob.glob(f'{AUDIO_PATH}/*')):
    uid = os.path.basename(i).split(".")[0]
    uid_out = os.path.join(TRANSCRIPT_PATH, uid)
    cmd = f"youtube-dl --all-sub --skip-download https://www.youtube.com/watch?v={uid} -o {uid_out}"
    os.system(cmd)
```

```python
for i,il in enumerate(glob.glob(f'{TRANSCRIPT_PATH}*.vtt')):
    l = os.path.basename(il).split('.')[1]
    ol = f"{os.path.join(TRANSCRIPT_PATH, l, os.path.basename(il))}"
    if not os.path.exists(os.path.dirname(ol)):
        os.makedirs(os.path.dirname(ol))
    cmd = f"mv {il} {ol}"
    os.system(cmd)
```

```python
def timetosec(t="00:00:48.340"):
    h, m, s = list(map(float, t.split(':')))
    sec = h*3600 + m*60 + s
    return sec

def transcriptfile2timestampslist(path):
    with open(path, 'r') as F:
        file = F.read()
    L = []
    for l in file.split('\n\n')[1:]:
        if not l: continue
        t = l.split('\n', 1)
        ts, tx = t
        tx = tx.replace('\n',' ').replace('<c>','').replace('</c>','').strip()
        tx = re.sub('<[0-9.:]+>','', tx)
        ts1, ts2 = ts.replace(" align:start position:0%","").split(' --> ')
        L.append([[timetosec(ts1), timetosec(ts2)], tx])
    return L

def audtrans2list(aud, T, sr=16000):
    L = []
    for l in T:
        t1 = int(sr*(l[0][0]-0.1))
        t2 = int(sr*l[0][1]+0.1)
        L.append([aud[t1:t2], l[1]])
    return L

def finalMaker(uid, AT):
    F = ""
    op = os.path.join(OUTPUT_PATH, uid)
    os.makedirs(op)
    for i, atline in enumerate(AT):
        F += f'{uid}_{str(i).zfill(6)}\t{atline[1]}\n'
        librosa.output.write_wav(os.path.join(op, f'{uid}_{str(i).zfill(6)}.mp3'), atline[0], sr=SAMPLERATE)
    with open(os.path.join(op, 'transcript.txt'), 'w') as file:
        file.write(F)


def ATmake(uid, audpath, transpath):
    T = transcriptfile2timestampslist(transpath)
    aud, sr = librosa.load(audpath, sr=16000)
    AT = audtrans2list(aud, T)
    finalMaker(uid, AT)
    return True
    
    
```

```python
ErrorFiles=[]
for i, audpath in enumerate(tqdm(glob.glob(f'{AUDIO_PATH}/*'))):
    uid = os.path.basename(audpath).split(".")[0]
    tpath = glob.glob(os.path.join(TRANSCRIPT_PATH, "en", f"{uid}*")) or \
            glob.glob(os.path.join(TRANSCRIPT_PATH, "auto", f"{uid}*"))
    tpath = tpath[0]
    print(uid, audpath, tpath)
    if os.path.exists(os.path.join(OUTPUT_PATH, uid)):
        print("\tSkipped")
        continue
    try:
        ATmake(uid, audpath, tpath)
    except Exception as e:
        print(f'\t\t ERROR with {uid}')
        ErrorFiles.append(uid)
print("Done Done Done")
print("Error with ", ErrorFiles)
```
